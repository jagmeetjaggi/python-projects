```python

def add_details(add_no):
    for i in range(add_no):
        year = int(input("\nEnter the year : "))
        print("\nEnter event details according to year: ", year)
        event_det = int(input("how much event you want to add : "))
        for k in range(event_det):
            event_name = input("- Event Name : ")
            country = input("- Country : ")
            athlete_name = input("- Athlete : ")
            medal = input("- Medal : ")
            #year = input("- Year : ")
            list1.append({
                "event_name": event_name,
                "country": country,
                "athlete_name": athlete_name,
                "medal" : medal,
                "year" : year
                })
    print(" \nDetails added successfull : ")



def veiw_details():
    print(f"{'Event Name':<20} {'Country':<15} {'Athlete Name':<20} {'Medal':<10} {'Year':<5}")
    print("-" * 70)
    for entry in list1:
        event_name = entry["event_name"]
        country = entry["country"]
        athlete_name = entry["athlete_name"]
        medal = entry["medal"]
        year = entry["year"]
        print(f"{event_name:<20} {country:<15} {athlete_name:<20} {medal:<10} {year:<5}")



def count_medal():
    medals_by_year = {}  

    for entry in list1:
        year = entry["year"]
        medal = entry["medal"]

        if year not in medals_by_year:
            medals_by_year[year] = {"gold": 0, "silver": 0, "bronze": 0}

        if medal == "gold":
            medals_by_year[year]["gold"] += 1
        elif medal == "silver":
            medals_by_year[year]["silver"] += 1
        elif medal == "bronze":
            medals_by_year[year]["bronze"] += 1

    print("\nTotal medals of Austria are ........")
    print(f"\n{'Gold':<20} {'Silver':<15} {'Bronze':<10} {'Year':<5} {'Count':<5}")
    print("-" * 70)

    for year, counts in medals_by_year.items():
        gold = counts["gold"]
        silver = counts["silver"]
        bronze = counts["bronze"]
        total_count = gold + silver + bronze
        print(f"{gold:<20} {silver:<15} {bronze:<10} {year:<5} {total_count:<5}")




def search_athlete():
    search = input("Search which athlete you want to search : ")
    #print(list1)
    print(f"{'Event Name':<20} {'Country':<15} {'Athlete Name':<20} {'Medal':<10} {'Year':<5}")
    for i in range(len(list1)):
        if(list1[i]["athlete_name"] == search):
            
            event_name = list1[i]["event_name"]
            country = list1[i]["country"]
            athlete_name = list1[i]["athlete_name"]
            medal = list1[i]["medal"]
            year = list1[i]["year"]
            print(f"{event_name:<20} {country:<15} {athlete_name:<20} {medal:<10} {year:<5}")



def switchfunc(no):
    match(no):
        case 1:
            add_no = int(input("Enter the no of year data you want to add : "))
            add_details(add_no)
        case 2:
            veiw_details()
        case 3:
            count_medal()
        case 4:
            search_athlete()
        case 5:
            print("Exit the program.....")
        case _:
            print(" Invalid choice")

def input_wrong(no):
    if(1 <= no <= 6):
        switchfunc(no)
    else:
        no = int(input("Enter the valid choice between 1 to 6 : "))
        input_wrong(no)

list1 = []
print("\n********** olympic details ************")
while(True):
    print("\n")
    print("Add Details 1 : ")
    print("Veiw Details 2 : ")
    print("veiw medal count by country 3 : ")
    print("Search Athelete Performance 4 : ")
    print("Exit 5 : ")
    
    no = int(input("\nEnter the choice :"))
    input_wrong(no)

```
